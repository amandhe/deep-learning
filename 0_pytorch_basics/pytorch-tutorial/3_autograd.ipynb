```python
import torch
```

```python
BATCH_SIZE = 16
DIM_IN = 1000
HIDDEN_SIZE = 100
DIM_OUT = 10

x = torch.randn(BATCH_SIZE, DIM_IN, requires_grad=False)
y = torch.randn(BATCH_SIZE, DIM_OUT, requires_grad=False)

class TinyModel(torch.nn.Module):
  def __init__(self):
    super().__init__()
    self.layer1 = torch.nn.Linear(DIM_IN, HIDDEN_SIZE)
    self.relu = torch.nn.ReLU()
    self.layer2 = torch.nn.Linear(HIDDEN_SIZE, DIM_OUT)

  def forward(self, x):
    x = self.layer1(x)
    x = self.relu(x)
    x = self.layer2(x)
    return x

model = TinyModel()
```

```python
print(model.layer1.weight[0][0:10])
print(model.layer1.weight.grad)
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.001)
prediction = model(x)
print(prediction.shape)
```

```python
loss = (y - prediction).pow(2).sum()
loss
```

```python
loss.backward()
print(model.layer1.weight[0][0:10])
print(model.layer1.weight.grad[0][0:10])
```

```python
optimizer.step()
print(model.layer1.weight[0][0:10])
print(model.layer1.weight.grad[0][0:10])
```

```python
for i in range(5):
  prediction = model(x)
  loss = (y - prediction).pow(2).sum()
  loss.backward()
  print(model.layer1.weight.grad[0][0:10])
```

```python
optimizer.zero_grad(set_to_none=False)
print(model.layer1.weight.grad[0][0:10])
```
